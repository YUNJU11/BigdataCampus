```python
station = pd.read_csv('seoul_station_address.csv')
station.head()
```

```python
print(station.info())
print(len(station[station['역위도'].isnull()]))
```

```python
null_station=station[station['역위도'].isnull()]
null_station
```

```python
station_name = []

for name in null_station['역사명']:
  station_name.append(name)

station_name
```

```python
station_address = []
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name, language='ko')
    station_address.append(tmp[0].get('formatted_address'))
    
    tmp_loc = tmp[0].get('geometry')
    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])


for address, lat, lng in zip(station_address, station_lat, station_lng):
    print('주소: ' + address)
    print('위도: ' + str(lat) + ' 경도: ' + str(lng))
    print('---------------------------------------')
```
